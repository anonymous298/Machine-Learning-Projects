# Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('matches.csv', index_col=0)
```

```python
data['team'].value_counts()
```

```python
data['round'].value_counts()
```

```python
data.dtypes
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['venue_code'] = data['venue'].astype('category').cat.codes
data['opp_code'] = data['opponent'].astype('category').cat.codes
data['hour'] = data['time'].str.replace(':.+','', regex=True)
data['day_code'] = data['date'].dt.dayofweek
```

```python
data['target'] = (data['result'] == 'W').astype(int)
```

```python
data
```

# Training our model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = data[data["date"] < '2022-01-01']
```

```python
test = data[data["date"] > '2022-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train['target'])
```

```python
pred = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test['target'], pred)
```
